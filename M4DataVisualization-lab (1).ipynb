# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Define the SQL query to retrieve the ConvertedComp data
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
#conn.close()

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()

```

### Box Plots


#Plot a box plot of `Age.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT Age
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Don't close the connection yet
# Plot the box plot for the 'Age' column
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=False)  # Drop NaN values for clean plotting
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Don't close the connection yet
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.6, color='blue')  # Scatter plot
plt.xlabel('Age')  # X-axis label
plt.ylabel('Work Week Hours')  # Y-axis label
plt.title('Scatter Plot of Age vs Work Week Hours')  # Title
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Don't close the connection yet
plt.figure(figsize=(8, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.6, color='pink')  # 's' sets bubble size

# Add labels and title
plt.xlabel('Work Week Hours')  # X-axis label
plt.ylabel('Code Revision Hours')  # Y-axis label
plt.title('Bubble Plot of Work Week Hours vs Code Revision Hours with Age as Bubble Size')  # Title

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
# your code goes here "DatabaseDesireNextYear"

QUERY = """
SELECT DatabaseDesireNextYear
FROM DatabaseDesireNextYear
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection after loading the data
conn.close()

# Count the occurrences of each database and get the top 5
db_counts = df['DatabaseDesireNextYear'].value_counts().head(5)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(db_counts, labels=db_counts.index, autopct='%1.1f%%', startangle=90)

# Add title
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')

# Display the pie chart
plt.show()
```

### Stacked Charts


print("Columns in DataFrame:", df.columns)Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
query = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(query, conn)
```

```python


#
median_workweekhrs = df['WorkWeekHrs'].median()
median_coderevhrs = df['CodeRevHrs'].median()

# Create a stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar('Age Group 30-35', median_workweekhrs, label='WorkWeekHrs', color='skyblue')
plt.bar('Age Group 30-35', median_coderevhrs, label='CodeRevHrs', color='lightgreen', bottom=median_workweekhrs)

# Add labels and title
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

query = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df = pd.read_sql_query(query, conn)

median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', color='b', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
query = """
SELECT MainBranch
FROM master
"""

df = pd.read_sql_query(query, conn)

mainbranch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue', edgecolor='black')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branch')
plt.show()

```
